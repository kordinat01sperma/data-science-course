```python
# Задание 17
P0 = np.array([[0, 0],[2, 2],[0, 1]])
P1 = np.array([[2, 0],[4, 2],[1, 3]])

P = np.array([[1, 1], [3, 0],[0, 3],[2, 3]])

# Шаг 1: Векторы отрезков
AB = P1 - P0
AB2 = np.sum(AB**2, axis=1, keepdims=True)

# Шаг 2: Векторы от начала отрезков до точек
AC = P[np.newaxis, :, :] - P0[:, np.newaxis, :]

# Шаг 3: Параметр t (проекция) для каждого отрезка и точки
t = np.sum(AC * AB[:, np.newaxis, :], axis=2) / AB2

# Шаг 4: Ограничиваем t от 0 до 1 (проекция на отрезок)
t_clipped = np.clip(t, 0, 1)

# Шаг 5: Ближайшая точка на отрезке
closest = P0[:, np.newaxis, :] + t_clipped[:, :, np.newaxis] * AB[:, np.newaxis, :]

# Шаг 6: Расстояние от исходной точки до ближайшей
diff = P[np.newaxis, :, :] - closest
distances = np.linalg.norm(diff, axis=2)

# Результат
print("\nМАТРИЦА РАССТОЯНИЙ (отрезки x точки):")
print("              ", end="")
for j in range(len(P)):
    print(f"Точка {j} ", end="")
print()
for i in range(len(P0)):
    print(f"Отрезок {i}: ", end="")
    for j in range(len(P)):
        print(f"{distances[i, j]:8.3f}", end="")
    print()
```

Допустим $X$ - зарплата, которую нам предлагают. Тогда $2X$ - увеличенная зарплата, $0,5X$ - уменьшенная зарплата.
