```python
# Вариант 4: Финансовый планировщик
print("=" * 40) 
print("Финансовый планировщик") 
print("=" * 40)

# Запрашиваем месячный доход
month_cost = float(input("Введите месячный доход в рублях: "))

# Запрашиваем какой процент от дохода будет отложен (0-100)
percent = float(input("Введите какой процент от дохода вы хотите откладывать (от 0 до 100): ")) / 100


# Рассчитаем сколько будет отложено за месяц, год, 5 лет
monthly_balance = month_cost - month_cost * percent
month_profit = month_cost * percent
year_profit = month_cost * 12 * percent
five_years_profit = month_cost * 12 * percent * 5


print(f"За месяц будет отложено - {month_profit} рублей")
print(f"За год будет отложено - {year_profit} рублей")
print(f"За 5 лет будет отложено - {five_years_profit} рублей")
print(f"Ежемесячно на жизнь остается - {monthly_balance} рублей")

# Добавим сложный ежегодный процент - 7%
year_percent = 0.07
i = 0
year_count = 0
hard_percent_profit = 0

while i < 120:
    hard_percent_profit_month = hard_percent_profit * year_percent / 12
    hard_percent_profit += month_profit + hard_percent_profit_month
    i += 1


ten_years_simple_profit = year_profit * 10
print(f"За 10 лет с учетом сложного процента будет отложено - {hard_percent_profit:.2f} рублей")
print(f"За 10 лет с учетом простого накопления будет отложено - {ten_years_simple_profit:.2f} рублей")

diff = hard_percent_profit -ten_years_simple_profit
print(f"Благодаря накоплению с процентами можно накопить на {diff} рублей больше за 10 лет")
```
